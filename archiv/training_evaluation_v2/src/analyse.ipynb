```python
import json
import matplotlib.pyplot as plt

def load_json(json_filename):
    with open(json_filename, "r") as json_file:
        json_data = json.load(json_file)
    return json_data["log"]
```

```python
denseNet121_log_4 = load_json("learning_path\denseNet121_0.0001_16_60_True_True_False_False.json")
denseNet121_log_5 = load_json("learning_path\denseNet121_1e-05_16_60_True_True_False_False.json")
denseNet121_log_6 = load_json("learning_path\denseNet121_1e-06_16_60_True_True_False_False.json")
denseNet121_log_7 = load_json("learning_path\denseNet121_1e-07_16_60_True_True_False_False.json")

val_accuracy_path = [
    denseNet121_log_4["val accuracy"],
    denseNet121_log_5["val accuracy"],
    denseNet121_log_6["val accuracy"],
    denseNet121_log_7["val accuracy"]
]

epoch = list(range(1, 61))
legend_labels = ["Lernrat 1e-4", "Lernrat 1e-5", "Lernrat 1e-6", "Lernrat 1e-7"]
plt.ylim(0.0, 1.0)

for i, data in enumerate(val_accuracy_path):
    plt.plot(epoch, data, label=legend_labels[i])

plt.title("Lernverlauf von DenseNet121")
plt.xlabel("Epoch")
plt.ylabel("Genauigkeit")

plt.legend()
plt.show()
```
